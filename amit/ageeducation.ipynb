```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python

houselisting = pd.read_csv("../Resources/age_education_housing.csv",encoding='ISO-8859-1')
houselisting
```

```python
plt.figure(figsize=(12, 10))
plt.barh(houselisting['AgeEducation'],houselisting['Edmonton'],color='blue')
plt.xlabel('HomeOwnership',fontsize=14)
plt.ylabel('Age&Education ',fontsize=14)
plt.title('Demograph of HomeOwnership in Edmonton ',fontsize=14)
plt.xticks(np.arange(start=0,stop=700000,step=50000)) 
plt.grid(axis="x")
plt.savefig("images/AgeEducationPopulation_Edmonton.png")
plt.show()
```
